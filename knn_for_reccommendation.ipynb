```python
import numpy as np
import pandas as pd
```

```python
courses_df = pd.read_csv("udemy_courses/Course_info.csv")
courses_df["id"] = courses_df["id"].astype("int64")
courses_df["num_subscribers"] = courses_df["num_subscribers"].astype("int64")
courses_df["content_length_min"] = courses_df["content_length_min"].astype("int64")
```

```python
courses_df
```

```python
def print_course_info(course_id, i):
    print("-"*40);
    if i == 0:
        print("----------TAKEN COURSE----------")
    elif i == 1:
        print("----------TOP RECOMMENDED COURSE----------")
    else:
        print("----------OTHER RECCOMMENDATIONS----------")
    print("-"*40);
    print("Course ID: {}".format(list(courses_df[courses_df["id"]==course_id]["id"])))
    print("Recommended Course: {}".format(list(courses_df[courses_df["id"]==course_id]["title"])))
    print("Instructor Name: {}".format(list(courses_df[courses_df["id"]==course_id]["instructor_name"])))
    print("Category: {}".format(list(courses_df[courses_df["id"]==course_id]["category"])))
    print("Sub-category: {}".format(list(courses_df[courses_df["id"]==course_id]["subcategory"])))
    print("Topic: {}".format(list(courses_df[courses_df["id"]==course_id]["topic"])))
    print("Number of Subscribers: {}".format(list(courses_df[courses_df["id"]==course_id]["num_subscribers"])))
    print("Average Rating: {}".format(list(courses_df[courses_df["id"]==course_id]["avg_rating"])))
    print("Content Length(min): {}".format(list(courses_df[courses_df["id"]==course_id]["content_length_min"])))
    print("Language: {}".format(list(courses_df[courses_df["id"]==course_id]["language"])))
    print("Price: {}".format(list(courses_df[courses_df["id"]==course_id]["price"])))
    print("-"*40)
```

```python
df = pd.read_csv("udemy_courses/reviews_cleaned.csv")
table = pd.pivot_table(df, values="rate", columns="user_id", index="course_id").fillna(0)
table
```

```python
from scipy.sparse import csr_matrix
```

```python
matrix = csr_matrix(table.values)
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
knn = NearestNeighbors(metric = "cosine")
knn.fit(matrix)
```

```python
course = np.random.choice(table.shape[0])
```

```python
dist, ind = knn.kneighbors(table.iloc[course].values.reshape(1, -1), n_neighbors=6)
```

```python
dist
```

```python
ind
```

```python
ind.flatten()
```

```python
for i in range(len(dist.flatten())):
    if i == 0:
        print_course_info(table.index[course], 0)
    else:
        print_course_info(table.index[ind.flatten()[i]], i)
```

```python
from sklearn.decomposition import TruncatedSVD
```

```python
svd = TruncatedSVD(n_components=12, random_state=1)
svd_matrix = svd.fit_transform(table.values)
```

```python
svd_matrix
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
corr = np.corrcoef(svd_matrix)
```

```python
course
```

```python
courses = list(table.index)
courses[course]
```

```python
courses
```

```python
indices = [i for i, x in enumerate(list(corr[course]>0.90)) if x == True]
indices
```

```python
corrs={}
for j in indices:
    corrs.update({j:list(corr[course])[j]})
```

```python
corr_dict = dict(reversed(sorted(corrs.items(), key=lambda item: item[1])))
```

```python
list(corr_dict)[:6]
```

```python
top_5 = [courses[index] for index in list(corr_dict)[:6]]
top_5
```

```python
for i in range(len(top_5)):
    if i == 0:
        print_course_info(top_5[i], 0)
    else:
        print_course_info(top_5[i], i)
```
