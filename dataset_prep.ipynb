```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("Comments.csv")
df = df.drop(columns=['id', 'date'])
df
```

```python
df = df.drop_duplicates()
df = df.dropna()
```

```python
reviewer_names_1 = df['display_name'].value_counts()
reviewer_names_1
```

```python
a = reviewer_names_1[(reviewer_names_1>3) & (reviewer_names_1<8)]
df = df.loc[df['display_name'].isin(a.axes[0])]
df
```

```python
z = df.groupby(["display_name"])
delete = []
for user in z.groups.keys():
     if z.get_group(user).nunique()["course_id"] != z.get_group(user).count()["course_id"]:
        delete.append(user)

df = df.loc[~df['display_name'].isin(delete)]
df.shape
```

```python
courses = df['course_id'].value_counts()
b = courses[(courses>75)]
df = df.loc[df['course_id'].isin(b.axes[0])]
df
```

```python
reviewer_names_2 = df['display_name'].value_counts()
reviewer_names_2
c = reviewer_names_2[reviewer_names_2>1]
df = df.loc[df['display_name'].isin(c.axes[0])]
df
```

```python
le = LabelEncoder()
reviewer_id = le.fit_transform(df["display_name"])
df.insert(1, "user_id", reviewer_id, True)
df
```

```python
df = df[["user_id", "display_name", "course_id", "rate"]]
df
```

```python
df.to_csv("reviews_cleaned.csv", index=False)
```
