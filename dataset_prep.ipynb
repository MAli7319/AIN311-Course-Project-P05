```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("Comments.csv")
df = df.drop(columns=['id', 'date', 'comment'])
df.shape
```

```python
df
```

```python
df = df.drop_duplicates()
df = df.dropna()
```

```python
reviewer_names = df['display_name'].value_counts()
```

```python
a = reviewer_names[(reviewer_names>10) & (reviewer_names<15)]
```

```python
a.axes[0]
```

```python
df = df.loc[df['display_name'].isin(a.axes[0])]
df
```

```python
courses = df['course_id'].value_counts()
courses
```

```python
b = courses[courses>50]
df = df.loc[df['course_id'].isin(b.axes[0])]
```

```python
le = LabelEncoder()
reviewer_id = le.fit_transform(df["display_name"])
df.insert(1, "user_id", reviewer_id, True)
df
```

```python
df = df[["user_id", "display_name", "course_id", "rate"]]
df
```

```python
df.to_csv("reviews_cleaned.csv", index=False)
```

```python
table = pd.pivot_table(df, values="rate", columns="course_id", index="user_id")
table.to_csv("utility_matrix.csv")
```
